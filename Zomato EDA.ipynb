```python
df_country=pd.read_excel('country-code.xlsx')
df_country.head()
```

```python
df.columns
```

```python
final_df=pd.merge(df,df_country,on='Country Code',how='left')
```

```python
final_df.head(2)
```

```python
## To check Data Types
final_df.dtypes
```

```python
final_df.columns
```

```python
country_names=final_df.Country.value_counts().index
```

```python
country_val=final_df.Country.value_counts().values
```

```python
## pie chart
plt.pie(country_val[:3],labels=country_names[:3],autopct='%1f%%')
```

Observation:Zomato maximum records or transcation are from i
India and after that USA and then United Kingdoms

```python
final_df.columns
```

```python
rating= final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
rating
```

## Observation 
1. When rating is b/w 4.5 to 4.9----> Excellent
2. When rating is b/w 4.0 to 3.4 ----> very good
3. When rating is b/w 3.5 to 3.9 ----> good
4. When rating is b/w 3.0 to 3.4 ----> average
5. When rating is b/w 2.5 to 2.9 ----> average
6. When rating is b/w 2.0 to 2.4 ----> poor

```python
rating.head()
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12,6)
sns.barplot(x="Aggregate rating",y="Rating Count",hue='Rating color', data=rating,palette=['blue','red','orange','yellow','green','green'])
```

Observation
1. Not Rated count is very high
2. Maximum number of rating are between 2.5 to 3.4

```python
## count plot
sns.countplot(x="Rating color", data=rating,palette=['blue','red','orange','yellow','green','green'])
```
